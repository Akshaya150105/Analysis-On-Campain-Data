```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=pd.read_csv('file.csv')
```

```python
df = pd.DataFrame(data)
customer_data = df[["Customer ID", "Region", "Age Group", "Income Group", "Purchase Frequency (Monthly)", "Avg. Spend (₹)"]]
campaign_data = df[["Campaign Type", "Campaign Cost (₹)", "Conversions (%)", "Total Revenue (₹)", "Population Estimate (Region)"]]
print("Customer Data:")
customer_data
```

```python
import seaborn as sns
customer_data['Total_Value'] = customer_data['Purchase Frequency (Monthly)'] * customer_data['Avg. Spend (₹)']

Avg_Group_analysis = customer_data.groupby('Age Group').agg({'Purchase Frequency (Monthly)': 'mean','Avg. Spend (₹)': 'mean','Total_Value': 'mean'
}).round(3)

income_group_analysis = customer_data.groupby('Income Group').agg({
    'Purchase Frequency (Monthly)': 'mean',
    'Avg. Spend (₹)': 'mean',
    'Total_Value': 'mean'
}).round(2)

segment_analysis = customer_data.groupby(['Age Group', 'Income Group'])['Total_Value'].mean().round(2)
most_profitable = segment_analysis.idxmax()

# Create visualization
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# Plot 1: Average Purchase Frequency by Age Group
sns.barplot(data=customer_data, x='Age Group', y='Purchase Frequency (Monthly)', ax=axes[0,0])
axes[0,0].set_title('Average Purchase Frequency by Age Group')
axes[0,0].tick_params(axis='x', rotation=45)

# Plot 2: Average Spend by Age Group
sns.barplot(data=customer_data, x='Age Group', y='Avg. Spend (₹)', ax=axes[0,1])
axes[0,1].set_title('Average Spend by Age Group')
axes[0,1].tick_params(axis='x', rotation=45)

# Plot 3: Average Purchase Frequency by Income Group
sns.barplot(data=customer_data, x='Income Group', y='Purchase Frequency (Monthly)', ax=axes[1,0])
axes[1,0].set_title('Average Purchase Frequency by Income Group')
axes[1,0].tick_params(axis='x', rotation=45)

# Plot 4: Average Spend by Income Group
sns.barplot(data=customer_data, x='Income Group', y='Avg. Spend (₹)', ax=axes[1,1])
axes[1,1].set_title('Average Spend by Income Group')
axes[1,1].tick_params(axis='x', rotation=45)

plt.tight_layout()

# Print analysis results
print("1. Age Group Analysis:")
print(Avg_Group_analysis)
print("\n2. Income Group Analysis:")
print(income_group_analysis)
print("\n3. Most Profitable Segment:")
print(f"Age Group: {most_profitable[0]}")
print(f"Income Group: {most_profitable[1]}")
print(f"Average Monthly Value: ₹{segment_analysis[most_profitable]:,.2f}")


plt.tight_layout()

# Show all plots
plt.show()
```

```python
# TASK 2

current_market_size = data.groupby('Region').agg({
    'Customer ID': 'count',
    'Total Revenue (₹)': 'sum'
}).reset_index()
current_market_size.columns = ['Region', 'Current Customers', 'Current Revenue']

population_data = data[['Region', 'Population Estimate (Region)']].drop_duplicates()
market_data = current_market_size.merge(population_data, on='Region')

market_data['TAM'] = (market_data['Current Revenue'] / market_data['Current Customers']) * market_data['Population Estimate (Region)']
market_data['TAM'] = market_data['TAM'].astype(int)

highest_growth_region = market_data.sort_values(by='TAM', ascending=False).iloc[0]

print("\nMarket Sizing:")
print(market_data)
print("\nRegion with the Highest Growth Potential:")
print(highest_growth_region)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(market_data['Region'], market_data['TAM'], color=['blue', 'green', 'red', 'purple'])
plt.xlabel('Region')
plt.ylabel('Total Addressable Market (₹)')
plt.title('Total Addressable Market per Region')
plt.show()
```

```python
grouped_df = data.groupby('Region').agg({
    'Purchase Frequency (Monthly)': 'mean',
    'Avg. Spend (₹)': 'mean',
    'Population Estimate (Region)': 'sum'
}).reset_index()

# Calculate Average Revenue per User (ARPU)
grouped_df['Average Revenue per User'] = grouped_df['Avg. Spend (₹)'] * grouped_df['Purchase Frequency (Monthly)']

# Calculate Total Potential Customers (Population Estimate)
grouped_df['Total Potential Customers'] = grouped_df['Population Estimate (Region)']

# Calculate TAM for each region
grouped_df['TAM'] = grouped_df['Average Revenue per User'] * grouped_df['Total Potential Customers']
grouped_df
```

```python
max_tam_region = grouped_df.loc[grouped_df['TAM'].idxmax()]

#TAM for each region and identify the highest growth potential region
print(grouped_df[['Region', 'TAM']])
print(f"\nRegion with the highest growth potential: {max_tam_region['Region']} with TAM: ₹{max_tam_region['TAM']:,.2f}")
```

```python
import matplotlib.pyplot as plt

# Visualizing TAM by region
plt.figure(figsize=(8, 6))
plt.bar(grouped_df['Region'], grouped_df['TAM'], color=['blue', 'green', 'orange', 'red'])

# Adding title and labels
plt.title('Total Addressable Market (TAM) by Region', fontsize=16)
plt.xlabel('Region', fontsize=12)
plt.ylabel('TAM (₹)', fontsize=12)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
#TASK 3

campaign_data['ROI (%)'] = ((data['Total Revenue (₹)'] - data['Campaign Cost (₹)']) /
                            (data['Campaign Cost (₹)']) )
campaign_data['ROI (%)']
ranked_campaigns = campaign_data.groupby('Campaign Type').agg({
    'ROI (%)': 'mean'
}).reset_index().sort_values(by='ROI (%)', ascending=False)

most_effective_campaign = ranked_campaigns.iloc[0]
least_effective_campaign = ranked_campaigns.iloc[-3:-1]

print("\nCampaign Analysis:")
print(ranked_campaigns)
print("\nMost Effective Campaign:")
print(most_effective_campaign)
print("\nLeast Effective Campaign:")
print(least_effective_campaign)
```

```python
plt.figure(figsize=(10, 6))
plt.barh(ranked_campaigns['Campaign Type'], ranked_campaigns['ROI (%)'], color='orange')
plt.xlabel('ROI (%)')
plt.ylabel('Campaign Type')
plt.title('ROI per Campaign Type')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

X = df[["Campaign Cost (₹)"]].values
y = df["Total Revenue (₹)"].values
model = LinearRegression()
model.fit(X, y)
intercept = model.intercept_
slope = model.coef_[0]
regression_eq = f"Revenue = {intercept:.2f} + {slope:.2f} * Campaign Cost"
new_campaign_costs = np.array([[100000], [200000], [300000]])
predicted_revenue = model.predict(new_campaign_costs)
```

```python
for cost, revenue in zip(new_campaign_costs.flatten(), predicted_revenue):
    print(f"Predicted Revenue for Campaign Cost ₹{cost}: ₹{revenue:.2f}")
```

```python
r2_score = model.score(X, y)
print(f"R-squared: {r2_score:.2f}")
```
